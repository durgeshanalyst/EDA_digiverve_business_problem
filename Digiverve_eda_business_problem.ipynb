```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```



1.   Problem Summary in short

*I understand. Building a model to forecast whether a client who already has health insurance with the company would also be interested in getting vehicle insurance from the same company is the issue at hand. By more effectively focusing on potential customers, the objective is to assist the organisation in maximising both its business plan and revenue. Demographic information, such as gender, age, and area code type, as well as information on the customer's car, policy, and premium are all included in the information that is available for analysis.*



*Generating Datasets with features*

1.   Age
2.   Gender
3.   region_code
4.   vehicle_age
5.   damage
6.   premium
7.   sourcing_channel
8.   vehicle_insurance





```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(123)

# Generate random data
data = {
    'gender': np.random.choice(['Male', 'Female'], size=1000),
    'age': np.random.randint(18, 75, size=10000),
    'region_code': np.random.choice(['ER', 'WR','NR', 'SR','CR','NER'], size=1000),
    'vehicle_age': np.random.randint(0, 10, size=1000),
    'damage': np.random.choice(['Yes', 'No'], size=1000),
    'premium': np.random.randint(1000, 5000, size=1000),
    'sourcing_channel': np.random.choice(['Online', 'Agent'], size=1000),
    'vehicle_insurance': np.random.choice([0, 1], size=1000)
}

# Create DataFrame
df = pd.DataFrame(data)

# Save DataFrame as Excel file
df.to_excel('random_dataset.xlsx', index=False)
```

```python
#Univariae Analysis
#Age
# Histogram of age distribution
sns.histplot(df['age'], bins=15,kde=True)
plt.xlabel('age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
```

```python
# Boxplot of age distribution
plt.boxplot(df['age'])
plt.ylabel('age')
plt.title('Boxplot of Age')
```

```python
# Barplot of gender distribution
sns.countplot(x='gender', data=df)
plt.xlabel('gender')
plt.ylabel('Count')
plt.title('Distribution of gender')
```

```python
# Barplot of region code distribution
sns.countplot(x='region_code', data=df)
plt.xlabel('Region Code')
plt.ylabel('Count')
plt.title('Distribution of Region Code')
plt.xticks(rotation=90)
```

```python
# Histogram of vehicle age distribution
plt.hist(df['vehicle_age'], bins=10)
plt.xlabel('Vehicle Age')
plt.ylabel('Frequency')
plt.title('Distribution of Vehicle Age')
```

```python
# Barplot of damage distribution
sns.countplot(x='damage', data=df)
plt.xlabel('Damage')
plt.ylabel('Count')
plt.title('Distribution of Damage')
```

```python
# Barplot of sourcing channel distribution
sns.countplot(x='sourcing_channel', data=df)
plt.xlabel('Sourcing Channel')
plt.ylabel('Count')
plt.title('Distribution of Sourcing Channel')
plt.xticks(rotation=90)
```

```python
# Barplot of vehicle insurance distribution
sns.countplot(x='vehicle_insurance', data=df)
plt.xlabel('Vehicle Insurance')
plt.ylabel('Count')
plt.title('Distribution of Vehicle Insurance')
```

```python
# Bivariate analysis with Age and Gender
sns.boxplot(x='gender', y='age', data=df)
```

```python
# Bivariate analysis with Age and Region Code
sns.violinplot(x='region_code', y='age', data=df)
```

```python
# Bivariate analysis with Vehicle Age and Damage
sns.scatterplot(x='vehicle_age', y='damage', data=df)
```

```python
# Bivariate analysis with Premium and Sourcing Channel
sns.barplot(x='sourcing_channel', y='premium', data=df)
```

```python
# Bivariate analysis with Vehicle Insurance and Gender
sns.countplot(x='vehicle_insurance', hue='gender', data=df)
```

```python
# Bivariate analysis with Premium and Vehicle Insurance
sns.histplot(x='premium', hue='vehicle_insurance', data=df, kde=True)
```

```python
# Multivariate analysis with Age, Gender, and Vehicle Insurance
sns.pairplot(df, vars=['age', 'gender', 'vehicle_insurance'])
```
